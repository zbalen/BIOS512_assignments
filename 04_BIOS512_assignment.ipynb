```r
library('tidyverse')
```

```r
turbines = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv')

turbines |> head()
```

Each row in the table represents a single turbine.

Make a bar chart üìä **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
turbines2 = turbines |>
    mutate(province_territory = province_territory |> fct_infreq())

turbines2 |> head()
```

```r
options(repr.plot.res = 250, repr.plot.width = 6, repr.plot.height = 6)
```

```r
p1 = ggplot(turbines2, aes(y = province_territory)) +
    geom_bar() + 
    labs(y = "Canadian Provinces", x = "Number of Turbines", title = "Canadian Wind Turbines") +
    theme_minimal() +
    theme(
        plot.title.position = 'panel',
        plot.title = element_text(color = '#242424', face = 'bold'))

p1
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
turbines3 = turbines2 |>
    mutate(province_territory = province_territory |> fct_lump(n = 5) |> fct_infreq() 
           |> fct_relevel('Other', after = 0))

turbines3 |> pull(province_territory) |> levels()
```

```r
p2 = ggplot(turbines3, aes(y = province_territory)) +
    geom_bar() +
    labs(y = "Canadian Provinces", x = "Number of Turbines", title = "Canadian Wind Turbines") +
    theme_minimal() +
    theme(
        plot.title.position = 'panel',
        plot.title = element_text(color = '#242424', face = 'bold'))

p2
```
